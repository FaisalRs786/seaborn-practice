# SEABORN PAIRPLOT

**SEABORN PAIRPLOT USING SNS.PAIRPLOT( DF )**

```python
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset('car_crashes').head(20)
df.head()
```

```python
sns.pairplot(df, vars = ['speeding','alcohol','not_distracted'], hue = 'abbrev', palette = 'Accent', kind = 'hist' , diag_kind= 'hist', markers = '*')
plt.show( )

# use vars parameter to select particular columns to create pairplot as
# hue_order we can set order of hue columns hue_order = []
# x_vars and y_vars are used to set how there columns names pairplots shown as :
# kind paramter is used to change kind of graph of  as kind = 'scatter', 'kde', 'hist','reg'
# diag_kind parameter is used to set kind of diagonal of graph as diag_kind = 'hist'
# markers is used to change markers in grapha as
```
