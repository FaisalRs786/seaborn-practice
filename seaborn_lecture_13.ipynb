# SEABORN CAT PLOT

**CAT PLOT IN SEABORN USING sns.catplot( x= '', y= '' , data = df)**

```python
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset('flights').head(20)
df
```

```python
sns.catplot(x = 'year', y = 'passengers', data = df, kind = 'box', height = 5)
plt.show( )

# kind parameter is used to set which kind of graph we want as kind = 'box' for box plot
# hue paramter is used to modify data according to other column as hue = 'month'
# palette paramter use to give color to the graph
# height use height to set height of graph as height  = 10
```
