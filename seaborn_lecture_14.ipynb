**STYLING DEABORN PLOT USING SNS.SET_STYLE( )**

```python
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset('exercise')
df
```

```python
sns.lineplot(x= 'diet', y = 'pulse', data = df, palette = 'BrBg')
sns.set_style('darkgrid')
plt.figure(figsize = (10,20))
sns.despine()
sns.set_context('paper',font_scale= 1)
plt.show( )

# sns.set_style('darkgrid') we can give 'whitegrid','dark','white'
# sns.dispine( ) to remove upper and lower side lines
# sns.figure(figsize = (10,20)) to  change graph size
# sns.set_context('paper',font_scale= 1)
```
