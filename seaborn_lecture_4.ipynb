# SEABORN BARPLOT

**SEABORN BAR PLOT using sns.barplt( x= '', y ='', data =df)**

```python
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset('mpg')
df
```

```python
sns.barplot( x ='cylinders', y = 'horsepower', data = df , hue = 'origin' , order = [8,6,4,3], hue_order = ['japan','usa','europe'],
            errorbar=('ci', 20), palette ='icefire',  saturation = 1)
sns.set(style = 'darkgrid')
plt.show( )


# hue parameter is used to modify data according to which column as gender example hue = 'gender'
# order parameter is used to set order of candel which candel comes first as order = [8,4,6]
# hue_order parameter is used to set order or hue column as our hue is origin in which have usa,europe and japan ser order as hue_order = ['usa','japan','europe']
# ci parameter is used to set err or bar as ci = 20 now use errorbar = ('ci', 20)
# n_boot = 2
# orient parameter is used to set graph horizental if both colmns are numbercal as orient = h
# color used to give color
# palette used to give color as palette = 'icefire'
# saturation paramter used to set saturation of graph as saturation = 0.5
# errcolor use to change color of err as errcolor  = 'r'
# errwidth used to set width of err as errwidth = 10
# capsize used to give cap to the bar as capsize = 0.5
# dodge parameter is used to merge candel bars as dodge = False
# alpha parameter is used to set transparency of graph as alpha = 0.5
```
