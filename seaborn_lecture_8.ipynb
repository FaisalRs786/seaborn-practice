# SEABORN VIOLIN PLOT

SEABORN VIOLIN PLOT AS sns.violinplot(x = '', y= '',data =df )

```python
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset('geyser').head(20)
df.head(10)
```

```python
sns.violinplot(x = 'kind', y = 'duration',data =df, hue = 'waiting', linewidth =2 ,palette  = 'Dark2',order=['short','long'] ,legend = False,
               saturation = 0.5, inner = 'quart',  )

plt.show()

# linewidth parameter is used to set width of line as linewidth = 2
# hue paramter
# color and palette parameter used to give color
# palette parameter is also used to give color as palette  = 'Dark2'
# order paramter used to set order to show data in graph as my x has data long and short i want show first short then long as order=['short','long']
# saturation used to give saturaiton of graph as saturation = 0.5
# split  =Ture
# scale parameter is used to scale graph accordin to width , area , count as  scale = 'area'
# inner parameter is used to change inner of voilin  of graph as inner = 'quart'
```
