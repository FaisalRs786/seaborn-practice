# COUNTPLOT USING SEABORN

COUNT PLOT USING SEABORN AS sns.countplot( x = ' ', data =df)

```python
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset('dots').head(100)
df
```

```python
sns.countplot(x = 'time', hue = 'align',data = df, palette = 'Accent', saturation = 0.5)

plt.show()
```
