# SEABORN STRIPLOT

**STRIP PLOT IN SEABORN USING sns.stripplot( x = '', y= '',data = df)**

```python
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset('dots')
df.head(  )
```

```python
sns.stripplot(x = 'coherence', y= 'time', data = df , hue = 'choice', palette = 'rocket_r', linewidth = 0.2, edgecolor = 'b',
              jitter = 0.2,size = 5, marker = '*' , alpha = 0.5)

plt.show( )

# hue paramter is used to ditribute data accordoing to column as hue = 'choice' which is T1 and T2
# palette parameter is used to give color to graph as palette = 'rocket_r'
# line_width paramter is used to set line width as  linewidth = 1
# edgecolor parameter is used to set egde color as edgecolor = 'r'
# jitter paramter is used to increase displacement or space between dots of data in stripplot as jitter = 0.2
# size paramter is used to set size of data dots as size = 5
# marker parameter is used to set marker shape as marker = '*'
# alpha paramter is used to set visibility of graph as alpha = 0.5
```

**single column stripplot**

```python
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset('dots')

sns.stripplot(y = df['time'])
plt.show()
```
