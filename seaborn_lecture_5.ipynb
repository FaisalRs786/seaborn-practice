# SCATTER PLOT

SCATTER BPLOT IN SEABORN using sns.scatterplot( x = "column1", y = "column2", data = df)

```python
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset("taxis").head(50)
df.head()
```

```python
sns.scatterplot(x = 'distance' , y ='tip' , data = df, hue = 'passengers',style = 'passengers', palette = 'gist_ncar', alpha =0.7)
plt.show()

# hue parameter used to modify data according to special colomun here is hue = 'passengers'
# style parameter is used to differnciate data of specific column as here is style = 'passengers'
# size paramter is used to differenciate coloumn according to size as size = 'passengers' if not in size
# pallete is used tpo give color as palatte = 'gist_ncar'
# alpha paramter is used to set tranparency as alpha =0.7
# marker paramter is used to change markers
```
