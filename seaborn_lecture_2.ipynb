# LINE PLOT IN SEABORN

**LINE PLOT IN SEABORN USING FUNCTION SNS.LINEPLOT( X = '', Y = '', DATA = DF )**

```python
# line plot in seaborn using function sns.lineplot(x= 'data frame column name ', y = ' data frame column name ', data= dataframe holder variable:)
import matplotlib.pyplot as plt
import seaborn as sns
df= sns.load_dataset('titanic').head(20) # use this function to load data from online source
df
```

```python
sns.lineplot(x = 'age', y = 'who', data = df, hue = 'alive', style = 'alive', palette = 'Accent', markers = ['<','o'], dashes = False)
plt.grid()
plt.show( )

# hue parameter used to give graph to distribute data according to this column
# palette  parasmter is used to give color to graph as : palette = 'Accent'
# style parameter is used to style differently to any column as style = 'gender'
# markers parameter is used to change style of markers as : markers = '<' or markers = ['o','>']
# dashes make  dashed False to make the line solid as dashes = False
# False the legend to remove the labels as legend = False
# plt.grid to give grid to the graph
# plt.title to give title to the graph
```
