# SEABORN BOXPLOT

**SEABORN BOX PLOT USING SNS.BOXPLOT( X = '' , Y = '' , DATA = DF)**

```python
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset('diamonds').head(20)
df
```

```python
sns.boxplot(x = 'cut', y = 'price', data = df, hue = 'color', order = ['Premium','Ideal','Fair','Good','Very Good'], showmeans = True,
            meanprops = {'marker':'<', 'markeredgecolor':'b'}, linewidth = 3,palette = 'Accent', )
plt.show( )

# hue paramter is used to modify data according to column
# color parameter is used to give color as
# order paramter is used to change order to show boxes as order = ['Premium','Ideal','Fair','Good','Very Good']
# showmeans parameter is used to show means points in boxes as showmeans = True
# meanprops paramter is used to give different properties to mean point as meanprops = {'marker':'<', 'markeredgecolor':'b'}
# linewidth = 3
# palette = 'Accent'
# orient 'h' if both x and y columns are numerical
```

**SINGLE COLUMN BOXPLOT**

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.boxplot(y = 'price', data = df)
plt.show()
```
