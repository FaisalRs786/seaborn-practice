# FACET GRID / MULTIPLE PLOT IN SEABORN

**FACE GRID / MULTIPLE PLOT IN SEABORN USING sns.facegrid(df, col ='', hue = '') and graph.map(plt.scatter).add_legend**

```python
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset('tips')
df.head()
```

```python
fig = sns.FacetGrid(df, col = 'day', hue = 'sex')
fig.map(plt.scatter, 'total_bill', 'tip').add_legend()
plt.show()
```
