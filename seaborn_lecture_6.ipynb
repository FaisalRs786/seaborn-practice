# SEABORN HEATMAP GRAPH

SEABORN HEATMAP GRAPH AS sns.heatmap(data=df)

```python
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset("titanic")
modified_df = df.drop(columns=['sex','embarked', 'class','who','adult_male','deck','embark_town','alive','alone']).head(10)
modified_df.head()
```

```python
sns.heatmap(modified_df, vmin = 0, vmax = 50, cmap = 'PuOr', annot = True , annot_kws= {'fontsize':10, 'color':'w'},
            linewidth = 1)
plt.show()

# vmin and vmax parameter is used to min and max value of heat map graph as vmin = 0, vmax = 50
# cmap parameter is used to change color as cmap = 'PuOr'
# annot paramter is used to show value at each color box as annot = True
# annot_kws is used to give different properties to the text of the heat map as annot_kws= {'fontsize:10, color:'w'}
# linewidth is used to give width between  each color box as linewidth = 2
# linecolor is used to give color to the width line between boxes as
# cbar used to remove color bar in graph as cbar = False
# xticklabel and yticklabel false to remove x and y labels from the graph
```
