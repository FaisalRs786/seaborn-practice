# HISTOGRAM PLOT IN SEABORN

HISTOGRAM PLOT USING SEABORN AS sns.displot(df['df dataset column name'])

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
df = sns.load_dataset('titanic').head(10)
df
```

```python
sns.displot( x= 'class', data= df, hue = 'sex', bins = [1,2,3,4,5] , palette = 'Accent', kde = True, rug = True, log_scale = True )
plt.show( )

# bins parameter are used to organize data in axis in pakiticular range as bins = [10,20,30,40,50,60]
# kde parameter is used to show kernal desnity estimator line in histogram chart as kde = True
# rug parameter is ussed to give rug to the bottom of the histogram chart as rug = True
# color parameter is used to give color to the chart as color= 'r'
# log_scale paramter is used to convert graph value in to log form as log_scale as log_scale = True
```
